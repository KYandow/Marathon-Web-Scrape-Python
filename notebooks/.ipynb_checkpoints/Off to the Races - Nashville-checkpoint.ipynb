```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def pull_nash_marathon_data(is_full=True, city='Nashville', year='2019'):
    full_in = '' if is_full else 'Half-'
    URL = 'https://www.runrocknroll.com/Events/' + city + '/The-Races/' + full_in + 'Marathon/' + year + '-Results'
    df_list = []
    row_shape, x = (1, 1)
    while row_shape > 0:    
        payload = {'page': str(x)}
        response = requests.post(URL, params=payload)        
        table_soup = bs(response.content, "html.parser").find(attrs={"class": "race-search-main"})
        df_page = pd.read_html(str(table_soup))[0]    
        row_shape = df_page.shape[0]
        df_list.append(df_page)
        x += 1   
    print('pulled ' + year + ' ' + str(is_full))
    return pd.concat(df_list)

##run pull function for Nashville races
fulls, halves = [], []
for yr in range(2016, 2020):
    fulls.append(pull_nash_marathon_data(is_full=True, city='Nashville', year=str(yr)))
    halves.append(pull_nash_marathon_data(is_full=False, city='Nashville', year=str(yr)))
print(str(len(fulls)) + ' ' + str(len(halves))
    
```

```python
for f in fulls:
    f['Race'] = 1
for h in halves:
    h['Race'] = 0.5
```

```python

all_lists = fulls + halves
all_races = pd.concat(all_lists, axis=0)
all_races.reset_index()
all_sorted = all_races.sort_values('Time', axis=0)
total_runner_years = all_sorted.shape[0]
```

```python
all_sorted.drop_duplicates(subset=['Name'], keep='first', inplace=True)
runners_deduped = all_sorted.shape[0]
```

```python
prop_duplicates = (total_runner_years - runners_deduped) / total_runner_years
print(round(prop_duplicates, 2))
## Average runner ran the race 1.27 times 1/(1 - 0.21)
```

```python
all_sorted['First Name'] = all_sorted['Name'].apply(lambda x: x.split(' ')[0])
print(all_sorted['First Name'])
```

```python
all_sorted['Last Name'] = all_sorted['Name'].apply(lambda x: x.split(' ')[-1])
print(all_sorted['Last Name'])
```

```python
less_dnf_sorted = all_sorted.loc[all_sorted['Time'] != '00:00:00']
```

```python
def pct_has_char(df, char_array, col):
    instances = 0
    for char in char_array:
        instances += df.loc[df[col].str.contains(char, regex=False)].shape[0]
    return round(100 * (instances / df.shape[0]), 2)

def count_char(name_string, lower_char):
    instances = 0
    for letter in name_string:
        if letter == lower_char or letter.lower() == lower_char:
            instances += 1
        else:
            pass
    return instances

other = ['ã', '³', 'é', '¹', 'ç', 'ž', 'ñ', 'Â', '€', '˜', 'â', '™', '£', 'í', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0']

##How many runners have hyphenated Last Names?
print('Percent Hyphenated ' + str(pct_has_char(less_dnf_sorted, ['-'], 'Name')))

## ##How many runners have apostrophied Last Names?
print('Percent apostrophied ' + str(pct_has_char(less_dnf_sorted, ["'", "`", "’"], 'Name')))

print('Percent parenthesized ' + str(pct_has_char(less_dnf_sorted, ['(', '{', '['], 'Name')))

print('Percent with periods ' + str(pct_has_char(less_dnf_sorted, ['.'], 'Name')))

print('Percent with other characters ' + str(pct_has_char(less_dnf_sorted, other, 'Name')))

```

```python
##Discover how many First Names are also the Last Names of another runner

is_last = less_dnf_sorted.loc[less_dnf_sorted['First Name'].isin(less_dnf_sorted['Last Name'])].shape[0]
print(less_dnf_sorted.loc[less_dnf_sorted['First Name'].isin(less_dnf_sorted['Last Name'])].head(50))
pct_is_last = 100 * (is_last / less_dnf_sorted.shape[0])
print(pct_is_last)
```

```python
def are_alpha(name_string):
    are_all_alpha = True
    for s in name_string:
        if (ord(s) >= 65 and ord(s) <= 90) or (ord(s) >= 97 and ord(s) <= 122) or (ord(s) == 32):
            pass
        else:
            are_all_alpha = False
            break
    return are_all_alpha

def spit_non_alpha(name_string):
    non_alphas = []
    for s in name_string:        
        if ((ord(s) >= 65 and ord(s) <= 90) or (ord(s) >= 97 and ord(s) <= 122) or (ord(s) == 32)):
            pass
        else:
            non_alphas.append(s)
    return non_alphas       
```

```python
non_alpha_names = less_dnf_sorted.loc[~less_dnf_sorted['Name'].apply(lambda x: are_alpha(x))]
print(non_alpha_names)
```

```python
less_dnf_copy = less_dnf_sorted.copy()
less_dnf_copy['A Count'] = less_dnf_sorted['Name'].apply(lambda x: count_char(x, 'a'))
less_dnf_copy['E Count'] = less_dnf_sorted['Name'].apply(lambda x: count_char(x, 'e'))
less_dnf_copy['I Count'] = less_dnf_sorted['Name'].apply(lambda x: count_char(x, 'i'))
less_dnf_copy['O Count'] = less_dnf_sorted['Name'].apply(lambda x: count_char(x, 'o'))
less_dnf_copy['U Count'] = less_dnf_sorted['Name'].apply(lambda x: count_char(x, 'u'))
less_dnf_copy['Total Characters'] = less_dnf_sorted['Name'].apply(lambda x: len(x))
```
